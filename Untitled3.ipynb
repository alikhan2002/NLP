```python
from docx import Document
```

```python

doc = Document('3.docx')
full_text = []
for paragraph in doc.paragraphs:
    full_text.append(paragraph.text)

print(full_text)
```

```python
import re
d = {}
curr = None
count = 43
full_text.append('end')
for s in full_text:
    if len(s):
        s = s.strip()
        if s == 'end':
            break
        if s[0].isdigit():
            cur = s
            p = r'\d+'
            match = re.search(p, cur)        
            if match:
                start, end = match.span()
                modified_string = cur[:start] + f'{count}' + cur[end:]
                count += 1
            cur = modified_string
            d[cur] = []
        else:
            d[cur].append(s)
print(d)
            
```

```python
ans = []
for x in d:
    s = x + '\n'
    s += '\n'.join(d[x])
    cnt = 0
    s = s.replace('\t', ' ')
    ans.append(s)
print(ans)
```
